## Approach 2  sliding window
**Idea:** <sliding window / hash map / two pointers / etc.>

**Time:** O(n)  |  **Space:** O(n)


```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        i = 0
        hash_table = {}
        max_len = 0
        for j in range(len(s)):
            if s[j] in hash_table.keys(): i = max(hash_table[s[j]]+1, i)
            current_len = j - i + 1
            max_len = current_len if current_len > max_len else max_len
            hash_table[s[j]] = j
        return max_len

# dvdfa    abcabcabc  tmmzuxt
```
