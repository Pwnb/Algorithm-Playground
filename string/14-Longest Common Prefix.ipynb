## Approach 1 — Brute Force
**Idea:** <explain in your own words>

**Time:** O(?)  |  **Space:** O(?)


```python
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        if len(strs) == 0: 
            return ''
        if len(strs) == 1: 
            return strs[0]
        def find_longest(str1, str2):
            res = []
            ite = min(len(str1), len(str2))
            for i in range(ite):
                if str1[i] == str2[i]:
                    res.append(str1[i])
                else:
                    break
            return ''.join(res)
        # print('res', find_longest("flower","flaaa"))
                    
        longest_str = find_longest(strs[0], strs[1])
        for i in range(len(strs)-1):
            l, r = strs[i], strs[i+1]
            temp_longest_str = find_longest(l, r)
            print(temp_longest_str)
            longest_str = temp_longest_str if len(temp_longest_str)<len(longest_str) else longest_str
        return longest_str
            


        
```

## Approach 2 — Optimized
**Idea:** <sliding window / hash map / two pointers / etc.>

**Time:** O(?)  |  **Space:** O(?)


```python
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        if len(strs) == 0: 
            return ''
        long_str = strs[0]
        for i in range(1, len(strs)):
            while strs[i].find(long_str) != 0:
                long_str = long_str[0: len(long_str)-1]
        return long_str
```
